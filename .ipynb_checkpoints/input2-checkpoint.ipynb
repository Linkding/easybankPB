```python
#多个列将txt文件中的列用zip转换成4*n的list
#strip()，可以去掉\n
x_0 = [x.strip().split(' ')[0] for x in open('bookvalue.txt').readlines()]
x_1 = [x.strip().split(' ')[1] for x in open('bookvalue.txt').readlines()]
x_2 = [x.strip().split(' ')[2] for x in open('bookvalue.txt').readlines()]
x_3 = [x.strip().split(' ')[3] for x in open('bookvalue.txt').readlines()]

frame = list(zip(x_0,x_1,x_2,x_3))
frame
```

```python
#转化成dataframe
df = pd.DataFrame(data = frame, columns=['codenumber', 'Date','bookvalue','changerate'])
df
```

```python
#保存为csv文件｀
df.to_csv('bookvalue3618.csv',index=False,header=True)
```

```python
#merge = stockdata.append(bookvalue,ignore_index=True)
#merge.sort_values('Date',ascending=True)
merge = pd.merge(stockdata, bookvalue, how='outer', on=['Date'])
#merge[merge.Date == '2013-06-30']
merge.sort_values('Date',ascending=False)

#成功将两个表格数据合并，并按时间顺序排列
```

```python
index_num = []
for i in bookvalue['Date']:
    h = merge[merge.Date == i].index.tolist()
    index_num.append(h)
#获取有净资产数据的列的index号   
```

```python
#上面tolist（）将index64转化为list，每个数值再通过切片获取，因为每个list里都只有一个元素
num1 = index_num[1][0]

merge['bookvalue'][num1]
#merge.loc[0:num1,'bookvalue'] = merge['bookvalue'][0:num1]
#merge.loc[0:num1,'bookvalue'] = 0.0
merge.loc[0:num1,'bookvalue'] = merge['bookvalue'][num1]
```

```python
for i in range(len(index_num)):
    i2 = i - 1
    num1 = index_num[i][0]
    num2 = index_num[i2][0]
    if i == 0:
        merge.loc[0:num1,'bookvalue'] = merge['bookvalue'][num1]
        merge.loc[0:num1,'changerate'] = merge['changerate'][num1]
    else:
        merge.loc[num2:num1,'bookvalue'] = merge['bookvalue'][num1]
        merge.loc[num2:num1,'changerate'] = merge['changerate'][num1]
```

```python
#计算出pb值
merge['PB'] = merge['Close'] /( merge['bookvalue'] / merge['changerate'])
merge[60:80]
del merge['codenumber']
```
